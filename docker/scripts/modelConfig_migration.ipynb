```python
import boto3
import os
import pandas as pd
import json

s3 = boto3.resource('s3')
s3_client = boto3.client('s3')
configPath = '_common/modelConfig.json'
```

```python
models = os.listdir('../../infra/models/environments/')
models = [name[:-7] for name in models]
model = models[0]
models
```

```python
def s3_exists(s3, bucket, key):
	try:
		s3.head_object(Bucket=bucket, Key=key)
		return True
	except:
		return False


# response = s3_client.get_object(Bucket=model, Key=configPath)
# response
```

```python
test = []
for model in models:
	bucket = s3.Bucket(model)
	print('-', model)
	if s3_exists(s3_client, model, configPath):
		print('exist')
		continue
	file = None
	for obj in bucket.objects.all():
		if 'attributesChoices.json' in obj.key:
			file = obj.key
			break
	print(file)
	if file is None:
		print('none')
		continue
	response = s3_client.get_object(Bucket=model, Key=file)
	json_data = json.loads(response['Body'].read().decode('utf-8'))
	config = {'version': 0, 'attributesChoices': {'links': {}, 'road_links': {}}, 'units': {}}
	config['attributesChoices']['links'] = json_data.get('pt', {})
	config['attributesChoices']['road_links'] = json_data.get('road', {})

	s3_client.put_object(
		Bucket=model,
		Key=configPath,
		Body=json.dumps(config).encode('utf-8'),
		ContentType='application/json',  # Sets correct file type in S3 browser
	)
	test.append(config)
```
